```python
from openai import OpenAI
api_key = ''
client = OpenAI(api_key=api_key)

completionIdeasList = client.chat.completions.create(
  model="gpt-4-0613",
  messages=[
    {"role": "system", "content": "You are a knowledgeable PhD in the field of Sociology, with an up-to-date and unbiased, vast knowledge of sociology."},
    {"role": "user", "content": "What are some of the top questions in Sociology? Give me 10 experiment ideas that I can test using surveys. Make sure the ideas are Ph.D. level and not trivial or over examined by the field already."}
  ]
)
print(completionIdeasList.choices[0].message.content)
ideas = completionIdeasList.choices[0].message.content
```

```python
completionExpandedIdea = client.chat.completions.create(
  model="gpt-4-0613",
  messages=[
    {"role": "system", "content": "You are a knowledgeable PhD in the field of Sociology, with an up-to-date and unbiased knowledge of the sociology."},
    {"role": "user", "content": "Given this list of topics, pick the most interesting one and expand on it. Write about things that you should consider to write good research questions and make note of interesting science on the topic. Pick only one as you will be designing a survey on it later. \n List of Topics: \n" + ideas}
  ]
)
print(completionExpandedIdea.choices[0].message.content)
expandedIdea = completionExpandedIdea.choices[0].message.content

completionJSON = client.chat.completions.create(
    model="gpt-4-turbo-preview",
    messages=[
        {"role": "system", "content": "You are a knowledgeable PhD in the field of Sociology, with an up-to-date and unbiased knowledge of sociology. Please generate responses in JSON format."},
        {"role": "user", 
         "content": f''' 
         You are given this topic and information:
         {expandedIdea}
         Write survey questions for a research analysis that can be taken by people on the internet to run this experiment. The questions should have a purpose and collect data that conclusions or novel insights can be drawn from. 
         Most importantly the questions should be in this JSON FORMAT ONLY so that they can be passed into the Google Forms API:
         {{
            "title": "Experiment Name",
            "questions": [
                {{"type": "MC", "title": "Experiment Question 1", "options": ["Excellent", "Good", "Average", "Poor"]}},
                {{"type": "TEXT", "title": "What do you think about Topic X?"}}
            ]
        }}
         Obviously, write as many questions as you need, usually around 2 demographic questions for age and ethnic background then 10-12 on the experiment will do. Although you may need more or less depending on the experiment and topic you were given. It's more likely that you will need to create multiple-choice questions as opposed to text but that decision is ultimately up to you. Make sure the questions are research-goal oriented and PhD level. You must return JSON and JSON only.
         '''
        }
    ],
    response_format={"type": "json_object"}
)

print(expandedIdea)
print("___________________")
print("messages")


print(completionJSON.choices[0].message.content)
jsonSurvey = completionJSON.choices[0].message.content
```

```python
import requests
import json
import re

# web app URL with the deployment ID
url = "https://script.google.com/macros/s/AKfycby08YLXwuaDDrRfjE_CSPLutky-AEnc7YFLgyHWVbfm6mATYTlduepB7-u7z0rQBFzN/exec"

# Data to be sent to Google Apps Script (form configuration)
data = json.loads(str(jsonSurvey))

# Sending a POST request to the Google Apps Script Web App
headers = {'Content-Type': 'application/json'}
response = requests.post(url, data=json.dumps(data), headers=headers)

# Printing the response from the web app (URL of the created form)
filename = re.sub(r'[^\w\s]', '', data['title']) + '.txt'
filename = filename.replace(' ', '_')  # Replace spaces with underscores for a cleaner filename

# Save the URL to a file named after the form title
with open(filename, 'w') as f:
    f.write(response.text)

# Print out the filename and URL for confirmation
print(f"Form URL saved in: {filename}")
print("URL:", response.text)
```
